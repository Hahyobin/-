```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import stft

input_folder = r'C:/Users/coops2022/시스템벤트/1-1. 윤활유 x (231204) A#3.5ss/1hz'
output_folder = r'C:/Users/coops2022/시스템벤트/spectrogram/윤활유X/1hz/x'

file_list = os.listdir(input_folder)

for file in file_list:
    if file.endswith('.csv'):
        file_path = os.path.join(input_folder, file)
        data = pd.read_csv(file_path)
        fs = 1
        nperseg = 256
        noverlap = 128
        nfft = 1024
        
        signal = data['x']
        f, t, Zxx = stft(signal, fs=fs, nperseg=nperseg, noverlap=noverlap, nfft=nfft)
        plt.figure()
        plt.pcolormesh(t, f, np.abs(Zxx), vmin=0, vmax=np.abs(Zxx).max(), shading='auto')
        
        plt.gca().set_axis_off()
        
        output_file_name = os.path.splitext(file)[0] + '.png'
        output_file_path = os.path.join(output_folder, output_file_name)
        plt.savefig(output_file_path)
        
        plt.close()
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft, resample

# 원본 샘플링 주파수 및 목표 샘플링 주파수
original_fs = 5000  # 원본 데이터의 샘플링 주파수
target_fs = 1       # 목표 샘플링 주파수

input_folder = r'C:/Users/coops2022/시스템벤트/1-2. 정상 (231204) A#3.5ss/1hz'
output_folder = r'C:/Users/coops2022/시스템벤트/spectrogram/정상/1hz/y'

file_list = os.listdir(input_folder)

for file in file_list:
    if file.endswith('.csv'):
        file_path = os.path.join(input_folder, file)
        data = pd.read_csv(file_path)

        # 원본 데이터
        signal = data['y']

        # 다운샘플링
        num_samples = int(len(signal) * target_fs / original_fs)
        resampled_signal = resample(signal, num_samples)

        # STFT 파라미터
        fs = target_fs
        nperseg = 256
        noverlap = 128
        nfft = 1024

        # STFT 수행
        f, t, Zxx = stft(resampled_signal, fs=fs, nperseg=nperseg, noverlap=noverlap, nfft=nfft)
        
        # 스펙트로그램 생성
        plt.figure()
        plt.pcolormesh(t, f, np.abs(Zxx), vmin=0, vmax=np.abs(Zxx).max(), shading='auto')
        plt.gca().set_axis_off()
        
        # 이미지 저장
        output_file_name = os.path.splitext(file)[0] + '.png'
        output_file_path = os.path.join(output_folder, output_file_name)
        plt.savefig(output_file_path)
        plt.close()
```
